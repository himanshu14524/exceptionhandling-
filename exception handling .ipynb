```python
Q1. What is an Exception in python ? Write the difference between exceptions and Syntax errors.

ANS--An exception in Python is an error that occurs during the execution of a program.
When a situation arises that disrupts the normal flow of the program, Python raises an 
exception to indicate that an error has occurred.

DIFFERENCE BETWEEN EXCEPTIONS AND SYNTAX ERRORS :
1.CAUSE 
Exceptions are raised at runtime when and error occurs during the excution of the program 
while syntax errors ,on the other hand , occur during the parsing (compilation) phase of the program .
2. TIMING :
Exceptions occurs at runtime when the specific line of code that triggers the exception is excuted while 
syntax errors are detected and reported by the python interpreter during the compilation (parsing) phase before the program 
starts executing .
```

```python
Q2. What happens when an exception is not handled ? Explain with an example .

ANS--When an exception is not handled, it leads to the termination of the program with an error
message (a traceback) that provides information about the unhandled exception and the point in the
code where it occurred. The traceback shows the sequence of function calls that led to the exception,
helping the developer identify the cause of the error.

#EXAMPLE 
def divide_numbers(a, b):
    result = a / b
    return result

try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = divide_numbers(num1, num2)
    print("Result of division:", result)
except ValueError as ve:
    print("ValueError: Invalid input. Please enter valid numbers.")
```

```python
Q3. Which python statements are used to catch and handled exceptions ? Explain with an example .

ANS--try, except, else, and finally statements are used to catch and handle exceptions. The try block
is used to enclose the code that might raise an exception, and the except block is used to handle the
exception if it occurs. The else block is optional and is executed if no exception occurs, and the finally
block is also optional and is executed regardless of whether an exception was raised or not.

#EXAMPLE

def divide_numbers(a, b):
    result = a / b
    return result

try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = divide_numbers(num1, num2)
    print("Result of division:", result)
except ValueError as ve:
    print("ValueError: Invalid input. Please enter valid numbers.")
except ZeroDivisionError as ze:
    print("ZeroDivisionError: Cannot divide by zero.")
else:
    print("Division successful.")
finally:
    print("This is the finally block. It will always execute.")
```

```python
Q4. Explain with an example :
    a. try and else 
    b. finally 
    c. raise 
    
ANS--
A. try and else :
The try and else blocks are used together to handle exceptions and execute code when no
exceptions occur. The code inside the try block is executed, and if an exception occurs, it jumps
to the corresponding except block (if present) to handle the exception. However, if no exception 
occurs in the try block, the code inside the else block is executed.

#EXAMPLE 
def divide_numbers(a, b):
    result = a / b
    return result

try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = divide_numbers(num1, num2)
except ValueError:
    print("ValueError: Invalid input. Please enter valid numbers.")
else:
    print("Result of division:", result)
    print("Division successful.")

B. finally:
The finally block is used to specify code that will be executed regardless of whether an exception
was raised or not. It allows you to perform cleanup operations or tasks that must be executed, regardless 
of the outcome of the try block.

#EXAMPLE 
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("ValueError: Invalid input. Please enter a valid number.")
finally:
    print("This block will always execute, regardless of exceptions.")

C. c. raise:
The raise statement is used to manually raise exceptions in Python. You can use it to raise built-in 
exceptions or create custom exceptions when certain conditions are met.

#EXAMPLE 
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    elif age < 18:
        raise ValueError("You must be 18 years or older.")

try:
    user_age = int(input("Enter your age: "))
    validate_age(user_age)
    print("Age is valid:", user_age)
except ValueError as ve:
    print(ve)
```
